```python
def f(t):
    return np.exp(-t)	*	np.cos(2*np.pi*t)

t1	=	np.arange(0.0,	20,	0.1)
t2	=	np.arange(0.0,	5.0,	0.02)
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(t1,f(t1))
plt.grid()
plt.subplot(212)
plt.plot(t2,f(t2))
plt.grid()
plt.show()
```

```python
plt.figure(1)
plt.subplot(221)
plt.plot(t1,np.sin(t1))
plt.title("the sine plot")
plt.subplot(222)
plt.plot(t1,np.cos(t1))
plt.title("the cosine plot")
plt.subplot(223)
plt.plot(t1,np.tan(t1))
plt.title("the tan plot")
plt.subplot(224)
plt.plot(t1,1/np.cos(t1))
plt.title("the tan plot")
```
